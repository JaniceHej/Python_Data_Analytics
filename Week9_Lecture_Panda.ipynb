```python
# import stuff
import pandas as pd
import numpy as np
import matplotlib as plt

## Pandas uses series and dataframe
## series = 2d array workings together
```

### Series

```python
# series
s1 = pd.Series([12,-4,88,9])
print(s1)
print("\n")
# to modify the index, do as below
s1 = pd.Series([12,-4,88,9], index=['a','b','c','d'])
print(s1)
print("\n")
s2 = pd.Series([12,-4,88,9], index=[1,2,3,4])
print(s2)
```

```python
s1.values
```

```python
s1.index
```

```python
# if index is changed to non-number, index = default
s1[2]
```

```python
# if index is changed to number, follow new index
s2[2]
```

```python
s1[0:2]
```

```python
for i in s1:
    if i <= 0:
        print(i)
```
