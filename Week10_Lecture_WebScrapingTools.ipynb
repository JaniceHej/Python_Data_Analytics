## Beautiful Soup

```python
import requests
import re        # regular expressions
import bs4
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

https://pythonspot.com/en/extract-links-from-webpage-beautifulsoup/

```python
# recall mainpage
mainpage = urllib.request.urlopen('https://archive.ics.uci.edu/ml/datasets/Forest+Fires')

# soup mainpage as sp
sp = soup(mainpage)
```

```python
# print sp
# show the results and what we want to scrap for
print(sp.prettify)
```

```python
# loop through sp with .findAll()
# findAll(name = None, attrs = {}, recursive = True, text = None, limit = None, ** kwargs)
#
# a = tagname
# attrs = {} Beautiful soup provides a special argument called attrs (dict that acts just like the keyword arguments)
# re.compile Is regex compile. we are telling it the line starts with "^" http://
#            The result is we tell the attrs to filter out href = "http://"

for link in sp.findAll('a', attrs = {'href' : re.compile("^http://")}):
    print(link)
```

```python
# using .get <---- returns value for the given key, in this case 'href'
# we can put this into a list .. etc
for link in sp.findAll('a', attrs = {'href' : re.compile("^http://")}):
    print(link.get('href'))
```
