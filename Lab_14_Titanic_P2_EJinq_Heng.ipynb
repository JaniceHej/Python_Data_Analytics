### Calculate the conditional probability that a person survives given their sex and passenger-class.

```python
g = sns.catplot(x="Pclass", hue="Sex", col="Survived", data=df, kind="count", height=4, aspect=.7);
```

### Probability for a female from different class to survive is as below:

Probability of a female from Class 1 to survive is 10.25%.

```python
df[(df['Pclass']==1)&(df['Survived']==1)&(df['Sex']=='female')]['Survived'].count()/df['Survived'].count()
```

Probability of a female from Class 2 to survive is 7.89%.

```python
df[(df['Pclass']==2)&(df['Survived']==1)&(df['Sex']=='female')]['Survived'].count()/df['Survived'].count()
```

Probability of a female from Class 3 to survive is 8.12%.

```python
df[(df['Pclass']==3)&(df['Survived']==1)&(df['Sex']=='female')]['Survived'].count()/df['Survived'].count()
```

### Probability for a male from different class to survive is as below:

Probability of a male from Class 1 to survive is 5.07%.

```python
df[(df['Pclass']==1)&(df['Survived']==1)&(df['Sex']=='male')]['Survived'].count()/df['Survived'].count()
```

Probability of a male from Class 2 to survive is 1.92%.

```python
df[(df['Pclass']==2)&(df['Survived']==1)&(df['Sex']=='male')]['Survived'].count()/df['Survived'].count()
```

Probability of a male from Class 3 to survive is 5.30%.

```python
df[(df['Pclass']==3)&(df['Survived']==1)&(df['Sex']=='male')]['Survived'].count()/df['Survived'].count()
```

#### In conclusion, a female will have a higher probability of surviving if she is in class 1, while a male has a higher probability of surviving if he is in class 3.
