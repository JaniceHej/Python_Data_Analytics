## Subplots 

```python
fig = plt.figure()
ax = fig.add_subplot(111) # 111 represents 1 tall, 1 wide, plot #1
ax.plot([1,2,3,4],[10,20,25,30], color='lightblue', linewidth=3)
ax.scatter([0.3,3.8,1.2,2.5],[11,25,9,26], color='red', marker='^', s=150)
#ax.set_xlm(0.5,4.5)
```

```python
fig2 = plt.figure()

# two sub plots
sp1 = fig2.add_subplot(121) # 1 wide 2 tall plot 1
sp2 = fig2.add_subplot(122) # 1 wide 2 tall plot 2

# looks and feels
fig2.suptitle('Even and odd') # main title

# plot some stuffs
# sp1
sp1.plot([2,4,6,8], color='blue', marker='o', linewidth=2)
sp1.set_title('even')
sp1.set_xlabel('x Even')
sp1.set_ylabel('Even Num')

# sp2
sp2.plot([3,5,7,9], color='red', marker='o', linewidth=2)
sp2.set_title('odd')
sp2.set_xlabel('x Odd')
sp2.set_ylabel('Odd Num')
```

```python
# increase or decrease the size of the image
fig2.set_size_inches(18.5, 10.5)
fig2.savefig('fig2-big.png', dpi=100)
```
