```python
# import
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# display the data
df = pd.DataFrame(pd.read_csv("iris.csv"))
df
```

```python
# print summary of the data
df.describe()
```

```python
# separate into sepal and petal datasets
sepalData = df[['sepal_length','sepal_width', 'species']]
petalData = df[['petal_length','petal_width', 'species']]
```

```python
# group by species
sepalByGroup = sepalData.groupby('species')
petalByGroup = petalData.groupby('species')
```

## Sepal length and width by species

```python
# print summary of the sepal data based on species
sepalByGroup.describe()
```

```python
# print sum of the sepal data based on species
sepalByGroup.sum()
```

```python
plt.style.use('ggplot')
sepalByGroup.sum().plot.bar(figsize=(20,10))
plt.title('Sepal Data')
plt.show()
```

## Petal length and width by species

```python
# print summary of the petal data based on species
petalByGroup.describe()
```

```python
# print sum of the petal data based on species
petalByGroup.sum()
```

```python
plt.style.use('ggplot')
petalByGroup.sum().plot.bar(figsize=(20,10))
plt.title('Petal Data')
plt.show()
```
