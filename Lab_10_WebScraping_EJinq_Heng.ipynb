```python
import urllib.request

# import beautiful soup
import requests
import re        # regular expressions
import bs4
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
# soup mainpage as sp
webpage = urllib.request.urlopen('https://archive.ics.uci.edu/ml/datasets/auto+mpg')
sp = soup(webpage)
```

```python
# show result of sp
print(sp.prettify)
```

```python
print("List of URLs: ")
# using .get to find all urls
for link in sp.findAll('a', attrs = {'href' : re.compile("^http://")}):
    print(link.get('href'))
```
